```python
import pandas as pd
import numpy as np
import ast
```

```python
df=pd.read_csv("anime.csv")
df
```

```python
new_df=df.drop("members",axis="columns")
new_df
```

```python
new_df=new_df.drop("episodes",axis="columns")
```

```python
new_df=new_df[["anime_id","name","genre","type","rating"]]
new_df
```

```python
new_df.iloc[0]
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.duplicated().sum()
```

```python
new_df.iloc[0].genre
```

```python
def to_list(val):
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return [val]  # just wrap it in a list if eval fails
```

```python
new_df['genre'] = new_df['genre'].apply(to_list)

```

```python
new_df.head()
```

```python
new_df['type'] = new_df['type'].apply(to_list)
new_df['rating'] = new_df['rating'].apply(to_list)
```

```python
new_df.head()
```

```python
new_df["tags"]=new_df["genre"] + new_df["type"] + new_df["rating"]
```

```python
new_df.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
new_df['tags'] = new_df['tags'].apply(convert3)
```
