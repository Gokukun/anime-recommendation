```python
import pandas as pd
import numpy as np
import ast
```

```python
df=pd.read_csv("anime.csv")
df
```

```python
new_df=df.drop("members",axis="columns")
new_df
```

```python
new_df=new_df.drop("episodes",axis="columns")
```

```python
new_df=new_df[["anime_id","name","genre","type","rating"]]
new_df
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.duplicated().sum()
```

```python
new_df.iloc[0].genre
```

```python
def convert(obj):
    if isinstance(obj, str):
        # Remove brackets and quotes if present
        obj = obj.strip("[]").replace("'", "").replace('"', "")
        # Split by comma
        return [x.strip() for x in obj.split(",")]
    elif isinstance(obj, list):
        # Already a list
        return obj
    else:
        raise ValueError("Unsupported input format")
```

```python
new_df['genre'] = new_df['genre'].apply(convert)
```

```python
new_df.head()
```

```python
new_df['genre'] = new_df['genre'].apply(lambda x:[i.replace(" ","_") for i in x])
```

```python
new_df
```

```python
new_df['tags'] = new_df['genre'].apply(lambda x: ', '.join(x)) + ' | ' + new_df['rating'].astype(str)

```

```python
new_df.head()
```

```python
anime = new_df[['anime_id','name','tags']]
```

```python
anime.head()
```

```python
anime["tags"] = anime["tags"].apply(lambda x:x.lower())
```

```python
anime.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
anime['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(anime['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
anime
```

```python
def recommend(movie):
    matches = anime[anime['name'] == movie]

    if matches.empty:
        print(f"'{movie}' not found in the dataset.")
        return

    # Get original index and position
    anime_index = matches.index[0]
    anime_pos = anime.index.get_loc(anime_index)

    # Use similarity based on position
    distances = similarity[anime_pos]
    anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    
    for i in anime_list:
        recommended_pos = i[0]
        recommended_index = anime.index[recommended_pos]
        print(anime.loc[recommended_index, 'name'])

```

```python
recommend("Under World")
```

```python
import joblib
joblib.dump(anime, 'anime.joblib')
joblib.dump(similarity, 'similarity.joblib')
joblib.dump(cv, 'vectorizer.joblib')
```
